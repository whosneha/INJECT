# Example Completeness Notebook

This notebook demonstrates how to compute and visualize completeness curves using the star cluster injection pipeline.

```python
# Import necessary modules
from src.data_access import load_coadd_image
from src.inject import inject_cluster
from src.completeness import compute_completeness_curve, plot_completeness
from src.io import save_catalog

# Load a coadded image
image = load_coadd_image('path/to/coadd/image')

# Inject clusters into the image
injected_clusters = []
for i in range(10):  # Inject 10 clusters
    cluster = inject_cluster(image, params={'brightness': 20, 'size': 5})
    injected_clusters.append(cluster)

# Compute completeness curve
completeness_curve = compute_completeness_curve(injected_clusters)

# Plot completeness curve
plot_completeness(completeness_curve)

# Save the injected catalog
save_catalog(injected_clusters, 'path/to/save/injected_catalog.h5')
```
